## Resampling -- Fraud

```python
from imblearn.under_sampling import RandomUnderSampler
import pandas as pd

X = combined.drop(['Fraud','FraudType'], axis=1)
y = combined['Fraud']
# Create a RandomUnderSampler instance
rus = RandomUnderSampler(random_state=42)

# Resample the data
X_resampled, y_resampled = rus.fit_resample(X, y)
combined_rus = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.Series(y_resampled, name='Fraud')], axis=1)
```

```python
combined_rus["Fraud"].value_counts()
```

```python
combined_rus
```

```python
from imblearn.over_sampling import RandomOverSampler
import pandas as pd

X = combined.drop(['Fraud','FraudType'], axis=1)
y = combined['Fraud']
# Create a RandomUnderSampler instance
rus = RandomOverSampler(random_state=42)

# Resample the data
X_resampled, y_resampled = rus.fit_resample(X, y)
combined_ros = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.Series(y_resampled, name='Fraud')], axis=1)
```

```python
combined_ros["Fraud"].value_counts()
```

```python
combined_ros
```
