## Resampling -- Fraud

```python
from imblearn.under_sampling import RandomUnderSampler
import pandas as pd

X = dmepos.drop(['Fraud','FraudType'], axis=1)
y = dmepos['Fraud']
# Create a RandomUnderSampler instance
rus = RandomUnderSampler(random_state=42)

# Resample the data
X_resampled, y_resampled = rus.fit_resample(X, y)
dmepos_rus = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.Series(y_resampled, name='Fraud')], axis=1)
```

```python
dmepos_rus["Fraud"].value_counts()
```

```python
dmepos_rus
```

```python
from imblearn.over_sampling import RandomOverSampler
import pandas as pd

X = dmepos.drop(['Fraud','FraudType'], axis=1)
y = dmepos['Fraud']
# Create a RandomUnderSampler instance
rus = RandomOverSampler(random_state=42)

# Resample the data
X_resampled, y_resampled = rus.fit_resample(X, y)
dmepos_ros = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.Series(y_resampled, name='Fraud')], axis=1)
```

```python
dmepos_ros["Fraud"].value_counts()
```

```python
dmepos_ros
```

```python
from imblearn.combine import SMOTEENN

X = dmepos.drop(['Fraud','FraudType'], axis=1)
y = dmepos['Fraud']
# Create a RandomUnderSampler instance
rus = SMOTEENN(random_state=42)

# Resample the data
X_resampled, y_resampled = rus.fit_resample(X, y)
dmepos_smote = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.Series(y_resampled, name='Fraud')], axis=1)
```

```python
dmepos_smote["Fraud"].value_counts()
```

```python
dmepos_smote
```
