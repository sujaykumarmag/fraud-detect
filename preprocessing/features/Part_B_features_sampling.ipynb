```python
val_partb = partb.sample(frac=0.1,random_state=0)
fraud_zero_indices = val_partb[val_partb["Fraud"] == 0].index
rows_to_remove = 100000

if len(fraud_zero_indices) >= rows_to_remove:
    random_indices_to_remove = np.random.choice(fraud_zero_indices, size=rows_to_remove, replace=False)
    val_partb = val_partb.drop(random_indices_to_remove)
else:
    print("Not enough rows with fraud=0 to remove.")
```

```python
val_partb["Fraud"].value_counts()
```

```python
fraud_zero_indices = val_partb.index
partb = partb.drop(fraud_zero_indices,axis=0)
```

## Sampling and save it  -- For Fraud

```python
from imblearn.under_sampling import RandomUnderSampler
import pandas as pd

X = partb.drop(['Fraud','FraudType'], axis=1)
y = partb['Fraud']
# Create a RandomUnderSampler instance
rus = RandomUnderSampler(random_state=42)

# Resample the data
X_resampled, y_resampled = rus.fit_resample(X, y)
partb_rus = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.Series(y_resampled, name='Fraud')], axis=1)
```

```python
partb_rus
```

```python
partb_rus["Fraud"].value_counts()
```

```python
from imblearn.over_sampling import RandomOverSampler
import pandas as pd

X = partb.drop(['Fraud','FraudType'], axis=1)
y = partb['Fraud']
# Create a RandomUnderSampler instance
rus = RandomOverSampler(random_state=42)

# Resample the data
X_resampled, y_resampled = rus.fit_resample(X, y)
partb_ros = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.Series(y_resampled, name='Fraud')], axis=1)
```

```python
partb_ros
```

```python
partb_ros["Fraud"].value_counts()
```

```python
from imblearn.combine import SMOTEENN

X = partb.drop(['Fraud','FraudType'], axis=1)
y = partb['Fraud']
# Create a RandomUnderSampler instance
rus = SMOTEENN(random_state=42)

# Resample the data
X_resampled, y_resampled = rus.fit_resample(X, y)
partb_smote = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.Series(y_resampled, name='Fraud')], axis=1)
```
